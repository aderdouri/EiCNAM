```python
from TorchAAD.Engine.stochastic_process import NormalProcess, LogNormalProcess
from TorchAAD.Engine.simulator import simulate_process
import matplotlib.pyplot as plt

# Simulate Lognormal Process
process = LogNormalProcess(mu=0.05, sigma=0.2)
paths = simulate_process(process, S0=100, T=1.0, steps=100, n_paths=10)

# Plot Results
for i in range(paths.shape[1]):
    plt.plot(paths[:, i].cpu().numpy())
plt.title("Lognormal Process Simulation")
plt.show()
```
